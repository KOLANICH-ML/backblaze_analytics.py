# Dependence of throughput from batch size

```python
import scipy as np
from matplotlib import pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
data=np.array((
	(1, 4.5),
	(2, 7.8),
	(3, 11),
	(4, 14),
	(5, 17),
	(10, 35),
	(15, 37),
	(20, 38),
	(30, 44),
	(40, 49),
	(50, 53),
	(60, 56),
	(70, 61),
	(80, 61),
	(90, 61),
	(100, 67),
	(200, 77),
	(300, 81),
	(500, 85),
	(900, 91),
	(2500, 95),
))
```

```python
plt.plot(data[:,0], data[:,1])
plt.xlabel("batch size (records)")
plt.ylabel("throughput (records/sec)")
plt.grid()
plt.show()
```
